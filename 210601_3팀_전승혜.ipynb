### 예제 1

```python
#예제 > 5개의 정수를 입력받아 합계 구하기
sum = 0 

for i in range(5):
    num = int(input('정수를 입력하세요: '))
    sum += num
print('합계출력:', sum)
```

### 예제 2

```python
# n개의 정수를 입력 받아 합계 출력
sum1 = 0
n = int(input('몇개의 숫자를 더할까요? :'))
for i in range(n):
    sum1 += int(input('정수입력:'))
print('합계출력:',sum1)
```

### 예제 3

```python
# n개의 정수를 입력 받아 짝수의 합계 구하기
sum1 = 0 
n = int(input('몇개의 숫자를 더할까요? :'))
for i in range(n):
    num = int(input('정수입력:'))
    if num % 2 == 0:
        sum1 += num

print('짝수의 합계:',sum1)
```

### 예제 4

```python
# 0이 입력될 때까지 숫자를 입력 받아 합계
sum1 = 0
num = 1   #그냥 ' '로 줘도 됨! 0만 아니면 됨.
while num != 0:
    num = int(input('숫자:'))
    sum1 += num
print('합계출력:',sum1)
```

###예제 5

```python
#1~n까지 더하는 프로그램을 작성하세요
sum2 = 0
n = int(input('1부터 몇까지 더할까요?:'))
for i in range(n):
    sum2 += i+1

print(f'1부터 {n}까지 더한 합계는: {sum2}')
```

```python
#숫자 n1, n2를 받아 n1~n2까지의 합계 출력
sum1 = 0
n1 = int(input('몇부터 더할까요?:'))
n2 = int(input('몇까지 더할까요?:'))
for i in range(n1, n2+1):   #끝값 포함 안하는거 명심
    sum1 += i

print(f'{n1}부터 {n2}까지 더한 합계는: {sum1}')
```

###예제 6

```python
#임의의 개수 숫자의 평균 구하기 - 개수가 정해져있을 때는 for문 사용
n = int(input('입력받을 정수의 개수:'))
sum1 = 0

for i in range(n):
    num = int(input('정수를 입력하세요: '))
    sum1 += num
ave = round(sum1/n, 2)
if n == 0:
    print('0개의 정수를 입력하셨습니다.')
else:
    print(f'평균: {ave}')
```

```python
#팩토리얼 계산하기 - 정수 n을 입력받아 n!을 계산하는 프로그램을 작성하세요.
n = int(input('몇 팩토리얼?: '))
factorial = 1
for i in range(n,0,-1):
    factorial *= i
print(f'{n}! = {factorial}' )
```

```python
#구구단
n = int(input('구구단 몇단을 출력할까요?: '))
for i in range(9):
    print(f'{n} x {i+1} = {n*(i+1)}', end = ' ')

#아니면
for i in range(1,10):     #시작값, 끝값을 활용하는 range도 많이 쓰자!
    print(f'{n} x {i} = {n*i}')
```

# 중첩반복구조

```python
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
#구구단 가로 출력
for i in range(2, 10):
    for j in range(1, 10):
        print(f'{i} x {j} = {i*j}', end = '\t')  #슬래시+t(ap): 줄바꿈 대신 4칸씩 떨어진다.
    print()
```

```python
#구구단 세로 출력
for i in range(1, 10):
    for j in range(2, 10):
        print(f'{j} x {i} = {i*j}', end = '\t')  #슬래시+t(ap): 줄바꿈 대신 4칸씩 떨어진다.
    print()
```

# 예제 7

```python
import random

ans = random.randint(1,100)

while True:
    num = int(input('정답은?: '))
    if ans < num:
        print('DOWN!')
    elif ans > num:
        print('UP!')
    else:
        print('정답!')
        break
```
