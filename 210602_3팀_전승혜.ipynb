###지역변수, 전역변수


1.  함수 내에서 만든 변수를 함수 밖에서 사용할 수 없다. (지역변수)
2.  함수 밖에서 만든 변수를 함수 내에서 사용할 수 있다. (전역변수)



* 함수 안에서 전역변수의 값을 변경하려면, global 키워드를 사용해야 한다.

```python
# 함수내 변수를 밖에서 사용 못함
def get_plus_minus(num1,num2):
    plus = num1 + num2
    minus = num1 - num2
    return plus,minus
# get_plus_minus(1,2)
print(plus,minus)
```

### 함수내 전역변수 변경 

```python
def circle_area():
    area = 3.14 * radius ** 2
    return area
area = 0
radius = float(input('반지름:'))
print('원의 넓이:',circle_area())
print(area) #변경안됨
```

```python
def circle_area():
    global area    #전역변수를 함수 내부와 연결해주는 역할
    area = 3.14 * radius ** 2
    return area
area = 0
radius = float(input('반지름:'))
print('원의 넓이:',circle_area())
print(area) #변경
```

```python
def circle_area():
    global area
    area = 3.14 * radius ** 2
    return area

area = 0 #변수의 초기값을 주는 습관을 들이자
radius = float(input('반지름: '))
print('원의넓이:', circle_area())
print(area)
```

**굵은 텍스트**### 매개변수 *args

```python
# 매개변수 개수 가변적일때 사용. 
def average(*args):     #*args는 튜플 데이터형
    avrg1 = sum(args) /len(args)
    return avrg1
print(average(1,2,3))
print(average(1,2,3,4,5))
```

```python
# 매개변수 개수 가변적일때 사용. 
def average(*a):
    avrg1 = sum(a) /len(a)
    return avrg1
print(average(1,2,3))
print(average(1,2,3,4,5))
```

##예제

```python
#소수 여부 판단하기
def is_prime(number):
    if number <2:
        return False
    for i in range(2, number):
        if number % i == 0:
            return False
        
    return True

i_num = int(input('숫자: '))
if is_prime(i_num) == True:
    print('소수입니다.')
else:
    print('소수가 아닙니다.')
```

##lambda()
* 간단한 함수일때 사용


```python
#lambda 함수 - 내부적인 지역함수 없이 한줄로 코드를 짤 수 있는 기술

def plus(num1, num2):
    result = num1+ num2
    return result
print(plus(10,20))

plus = lambda num1, num2: num1+num2
print(plus(10,20))
```

##map()
* 두개의 인자를 갖는 함수로 순서가 있는 자료형에 사용
* 첫번째 인자는 함수명, 두번째인자는 순서가 있는 자료형(ex. 리스트)

```python
#map 함수
mylist = [1,2,3,4,5]
plus10 = lambda num: num+10

mylist = list(map(plus10, mylist))
print(mylist)
```

##재귀함수
알고리즘을 구현할 때 유용하다

```python
def hello(count):
    if count == 0: #반드시 종료조건을 만들어야 한다. 
        return

    print('hello world!', count)

    count -= 1    # count를 1감소시킨 뒤 
    hello(count)   # 다시 hello에 넣음

hello(5)
```

```python
#예제> 팩토리얼 만들기

def factorial(n):
    if n == 1:
        return 1

    return n * factorial(n-1)

print(factorial(6))
```

###예제 - 편의점 재고 관리

```python
#초기재고 설정
inventory = {'삼각김밥':10, '커피우유':10}

#1. 재고등록 함수

def add_product():
    while True:

        product = input('상품명: ')
        if product == '':
            break
        num = int(input('개수: '))

        if product in inventory.keys():
            inventory[product] += num
            #inventory.update(product=num)

        else:
            inventory[product] = num

    print('-'*50)
    for key, value in inventory.items():
        print(f'{key} : {value}')

#2. 제품판매 함수

def sell_product():
    while True:

        product = input('상품명: ')
        if product == '':
            break
        num = int(input('개수: '))

        if product in inventory.keys():
            if inventory[product] >= num:
                inventory[product] -= num
                print(inventory[product], '개 남았습니다.')
            else:
                print('재고가 부족합니다.')

        else:
            print('판매하지 않는 상품입니다.')


#3. 재고조회 함수
def search_product():
    while True:

        product = input('상품명: ')
        if product == '':
            break
        print(inventory.get(product,'등록되지 않은 상품'))
        num = inventory[product]
        print(num, '개 남았습니다.')


# 시스템 설정

print('''========== 재고관리 시스템 ==========
1: 재고등록, 2: 제품판매, 3: 재고확인, 0: 프로그램 종료
''')

while True:
    menu = input('메뉴를 선택하세요: ')
    if menu == '1':
        add_product()
    elif menu == '2':
        sell_product()
    elif menu == '3':
        search_product()
    elif menu == '0':
        print('프로그램을 종료합니다.')
        break
        
```
